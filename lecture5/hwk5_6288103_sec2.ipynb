```python
import numpy
```

```python
def fmax(x):
    return x*(1-x)*numpy.e**x
```

```python
def fmin(x):
    return x**5 - 3*x**4 + 5
```

```python
def max_interval(f,a,b,p):
    ep = (b-a)/1000
    ea = 100
    es = 0.5 * 10**(2-p)
    px = b
    while ea > es:
        l = (a+b)/2 - ep/2
        u = (a+b)/2 + ep/2
        if f(l) > f(u):
            b = u
        else:
            a = l
        xmax = (a+b)/2
        ea = abs((xmax - px)/xmax) * 100
        px = xmax
        print('({},{}) xmax = {} ea = {}'.format(a,b,xmax,ea))
```

```python
max_interval(fmax,0,2,4)
```

```python
def min_goldensearch(f,xL,xU,p):
    phi = (1+numpy.sqrt(5))/2 
    ea = 100
    es = 0.5 * 10**(2-p)
    while ea > es:
        d = (phi - 1)*(xU-xL)
        x1 = xL+d
        x2 = xU-d
        if f(x1) > f(x2):
            xmin = x2
            xU = x1
        else:
            xmin = x1
            xL = x2
        ea = (2-phi)*abs((xU-xL)/xmin)*100
        print('({},{}) xmin = {} ea = {}'.format(xL,xU,xmin,ea))
```

```python
min_goldensearch(fmin,0,4,4)
```
